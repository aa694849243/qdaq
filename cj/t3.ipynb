```python
from numba import njit, vectorize, prange, guvectorize, float64, types, int64,objmode
import numpy as np
from numba.typed import Dict
import math
```

```python
lst = np.random.randn(1000)+100
dct = {'xi2': lst}


@njit(nogil=True, fastmath=True)
def oned_rms(temptd, calc_size):
    return np.sqrt(np.sum(temptd['xi2']) / (calc_size * len(temptd['xi2'])))


@guvectorize([(float64[:], int64, float64)], '(n),()->()')
def oned_rms3(lst, calsize, res):
    # res = 0
    # for num in lst:
    #     res += num
    # res = res / (calsize * len(lst))
    with objmode(res='types.float64'):
        res=math.sqrt(np.sum(lst)/(calsize*len(lst)))


def oned_rms2(temptd, calc_size):
    return np.sqrt(np.sum(temptd['xi2']) / (calc_size * len(temptd['xi2'])))
```

```python
ndct = Dict.empty(key_type=types.unicode_type, value_type=types.float64[:])
ndct['xi2'] = lst
```

```python
print(oned_rms(ndct, 100), oned_rms2(ndct, 100), oned_rms3(lst, 100))
```
